# 3 Testes de uma amostra
## 3.1 O efeito do tamanho de uma amostra

```python
# Queremos analisar como a média das notas médias se comporta conforme mais dados são atribuídos
media = []
s = 0
temp = media_filmes_mais_de_10_notas.sample(frac=1, random_state=75243)
for i in range(0, len(temp)):
    s = temp[0:i+1].mean()
    media.append(s)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(media)
ax.set_title('Média das Notas conforme o aumento de dados', fontsize=15)
ax.set_xlabel('Quantidade de registros', fontsize=13)
ax.set_ylabel('Média calculada', fontsize=13)
```

A Média vai se tornando mais consistente conforme aumentamos a nossa amostra
Agora vamos calcular o intervalo de confiança para os testes z e t, respectivamente. 95% de confiança
    https://www.statsmodels.org/dev/generated/statsmodels.stats.weightstats.zconfint.html
    https://www.statsmodels.org/dev/generated/statsmodels.stats.weightstats.DescrStatsW

```python
from statsmodels.stats.weightstats import zconfint

zconfint(media_filmes_mais_de_10_notas, alpha=0.05)
```

```python
from statsmodels.stats.weightstats import DescrStatsW

desc_todos_com_10_votos = DescrStatsW(media_filmes_mais_de_10_notas)
desc_todos_com_10_votos.tconfint_mean(alpha=0.05)
```

# 4 Outros testes
## 4.1 ZTest para uma amostra
Vamos comparar a média de um filme em específico Toy Story

```python
movie = pd.read_csv('dados/movies.csv')
movie.head()
```

```python
notas1 = notas.query('movieId == 1')['rating']
notas1
```

```python
ax = sns.histplot(notas1, kde=True, stat='density')
ax.figure.set_size_inches(12,8)
ax.set_xlabel('Nota', fontsize=13)
ax.set_ylabel('Densidade', fontsize=13)
```

```python
ax = sns.boxplot(data= notas1, orient='h')
ax.figure.set_size_inches(12,8)
ax.set_xlabel('Nota', fontsize=13)
ax
```

```python
print(f'Média dos filmes = {media_filmes_mais_de_10_notas.mean()} \nMédia Toy Story = {notas1.mean()}')
```

```python
from statsmodels.stats.weightstats import ztest
#Teste para verificar como a média toy story pode ser equivalente a media dos filmes. Resultado abaixo de 0.05 pode ser descartado a hipótese nula

print(f'p-value = {ztest(notas1, value=media_filmes_mais_de_10_notas.mean())[1]} ')
```

```python
import numpy as np
```

```python
np.random.seed(75241)
temp = notas1.sample(frac=1)


def calcula_teste(i):
    media = temp[0:i].mean()
    stat, p = ztest(temp[0:i], value= 3.4320503405352594)
    return(i, media, p)

media = np.array([calcula_teste(i) for i in range(2,len(temp))])
#Transformamos em um array numpy para podermos fatia-lo   
    
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(media[:,0], media[:,1])
ax.set_title('Medias Toy Story aumentando os registros', fontsize=15)
ax.set_xlabel('Quantidade de Registros', fontsize=13)
ax.set_ylabel('Média', fontsize=13)
ax.plot(media[:,0] , media[:,2])
ax.axhline(y=0.05, color='r', linestyle='--')
```

Acima plotamos os valores da média e o resultado do testz (para a hipótese de que a média ToyStory é a mesma da média dos filmes gerais) conforme aumentamos nossa amostra. Repare pelo tracejado vermelho (95% confiança) que para pequenas quantidades de amostras poderíamos obter o resultado inválido de que nossa hipótese estaria certa. Na verdade obteríamos o resultado de que não poderíamos descartar a hipótese com a quantidade de amostra fornecida. 

# 05 Testes para duas amostras
## 5.1 Comparando a média de duas amostras

```python
from statsmodels.stats.weightstats import zconfint
print(f'Intervalo de confiança com 95% de confiança para a média de Toy Story {zconfint(notas1)}')
```

```python
print(f'Usando o zconfint c/ 2 distribuições (Notas de Toy Story e Notas de todos os filmes), podemos intervalar o quanto a média de um se compara com a média do outro. \nOu seja a média de Toy Story está acima em {zconfint(x1=notas1, x2=notas["rating"])} unidades da média de todos os filmes (95% de confiança), de acordo com o ztest)')
```

```python
zconfint(x1=notas['rating'], x2=notas1)
#Média de todos os filmes comparada com a média de Toy Story, ou seja, o oposto do resultado acima
```

```python
print(f'Para as duas distribuições temos p-value = {ztest(x1=notas1, x2=notas["rating"])[1]}')
```

```python
from scipy.stats import ttest_ind

ttest_ind(a=notas1, b=notas['rating'])
#t_test para calcular o p-value de duas amostras
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html
```

    https://www.statsmodels.org/dev/generated/statsmodels.stats.weightstats.DescrStatsW
    https://www.statsmodels.org/dev/generated/statsmodels.stats.weightstats.DescrStatsW.get_compare.html#statsmodels.stats.weightstats.DescrStatsW.get_compare
    https://www.statsmodels.org/dev/generated/statsmodels.stats.weightstats.CompareMeans.html#statsmodels.stats.weightstats.CompareMeans

```python
from statsmodels.stats.weightstats import DescrStatsW

descr_todas_as_notas = DescrStatsW(notas['rating'])
descr_toystory = DescrStatsW(notas1)
comparacao = descr_todas_as_notas.get_compare(descr_toystory)
comparacao.summary(use_t=True)
# Usando statsmodels para obter as descrições do t-test
```

O intervalo de confiança entre as duas amostras, mostrando a comparação com relação as médias. o p-value ínfimo que nos leva a descartar a hipótese nula são bem parecidos com o ztest algumas células acima.

```python
comparacao.summary(use_t=False) #usando o ztest
```

## 5.3 Cuidado ao interpretar o Intervalo de Confiança

```python
fig, ax = plt.subplots(figsize=(12,8))
ax = sns.boxplot(data=[notas['rating'], notas1[3:12]], orient='h') # Poucos dados 'aleatoriamente'
ax.set_title('Notas Médias - Box Plot', fontsize=16)
ax.set_xlabel('Nota Média', fontsize=15)
ax.set_yticks(ticks=[0,1], labels=['Filmes Gerais', 'Toy Story'], fontsize=14)
ax
```

```python
# Usando o z_test, o que não deveríamos devido ao baixo número de amostras, para testarmos a hipótese nula
descr_toystory = DescrStatsW(notas1[3:12])
descr_todas_as_notas = DescrStatsW(notas['rating'])
comparacao = descr_todas_as_notas.get_compare(descr_toystory)
comparacao.summary(use_t=False)
```

Nosso p-value foi maior que 0.05 - logo não temos como descartar a hipótese nula (devido a falta de dados da amostra)


Nosso intervalo de confiança varia em torno do 0, ou seja, não nos diz nada o quanto a média dos filmes gerais se compara com relação a média de Toy Story

Suponha agora um IC = (-0.2, -0.1). Ou seja praticamente nulo e portanto as médias são praticamente iguais, como por exemplo a média das notas de um filme. Entretando se estivermos comparando, por exemplo, a taxa de sobrevivência entre dois remédios distintos, temos uma diferença importante para ser considerada.
