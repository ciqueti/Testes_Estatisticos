# 01 Conhecendo os dados e explorando
## 1.1 Limpando os dados e visualizando uma distribuição

```python
import pandas as pd


tmdb = pd.read_csv('dados/tmdb_5000_movies.csv')
tmdb.head()
```

```python
tmdb.describe()
```

```python
import seaborn as sns

sns.set_theme(style='darkgrid', palette='tab10')
```

```python
ax = sns.histplot(tmdb['vote_average'], kde=True, stat='density')
ax.figure.set_size_inches(14,8)
ax.set_ylabel('Frequência', fontsize=13)
ax.set_title('Distribuição das notas médias', fontsize=15)
ax.set_xlabel('Nota Média', fontsize=13)
```

```python
ax = sns.boxplot(data=tmdb, x='vote_average')
ax.figure.set_size_inches(12,8)
ax.set_title('Distribuição das Notas Médias do TMDB 5000', fontsize=15)
ax.set_xlabel('Nota Média', fontsize=15)
```

Notas Médias extremas como 0 e 10 são suspeitas, vamos analisar

```python
tmdb.query('vote_average == 0').head(10)
```

```python
tmdb_mais_10_votos = tmdb.query('vote_count >= 10')
tmdb_mais_10_votos.describe()
```

```python
ax = sns.histplot(tmdb_mais_10_votos['vote_average'], kde=True)
ax.figure.set_size_inches(12,8)
ax.set_title('Distribuição das Notas Médias do TMDB 5000 (acima de 10 notas)', fontsize=15)
ax.set_xlabel('Nota Média', fontsize=15)
```

```python
ax = sns.boxplot(data=tmdb_mais_10_votos, x='vote_average')
ax.figure.set_size_inches(12,8)
ax.set_title('Distribuição Notas Médias (acima de 10 notas)', fontsize=15)
ax.set_xlabel('Nota Média', fontsize=13)
```

## 1.2 Limpando e Visualizando o movielens

```python
# Colecionando média das notas por filme
notas = pd.read_csv('dados/ratings.csv')
media_filmes = notas.groupby(by='movieId').mean()
media_filmes = media_filmes['rating']
media_filmes
```

```python
ax = sns.histplot(media_filmes, kde=True, )
ax.figure.set_size_inches(12,8)
ax.set_title('Distribuição Nota Média Movielens', fontsize=15)
ax.set_xlabel('Nota Média', fontsize=12)
```

```python
# Vamos seleceionar os filmes que tiveram pelo menos 10 notas
movie_Id = (notas.groupby(by='movieId').count()).query('rating >= 10').index
movie_Id.values
```

```python
media_filmes_mais_de_10_notas = media_filmes.loc[movie_Id.values]
```

```python
ax = sns.histplot(data=media_filmes_mais_de_10_notas, kde=True)
ax.figure.set_size_inches(12,8)
ax.set_title('Média Filme com mais de 10 notas - movielens', fontsize=15)
ax.set_xlabel('Média do filme', fontsize=13)
ax.set_ylabel('Frequência', fontsize=13)
```

```python
ax = sns.boxplot(data=media_filmes_mais_de_10_notas.values, color='g', orient='h')
ax.figure.set_size_inches(12,8)
ax.set_title('Média Filmes com mais de 10 notas - movielens', fontsize=15)
ax.set_xlabel('Média do Filme', fontsize=13)
```

## 1.3 Função de Distribuição acumulada
Dado um filme, Toy Story nota media 3.7, o quão bom ou o quão acima dos outros filmes ele está?
Para responder isso podemos usar a distribuição acumulada

```python
ax = sns.histplot(data=media_filmes_mais_de_10_notas.values, cumulative=True, kde=True, stat='density')
ax.figure.set_size_inches(12,8)
ax.set_title('Distribuição Acumulada - Nota Média Filmes movielens', fontsize=15)
ax.set_xlabel('Nota Média', fontsize=13)
ax.set_ylabel('Densidade', fontsize=13)
```

```python
ax = sns.histplot(data=tmdb_mais_10_votos, x='vote_average', stat='density', kde=True, cumulative=True)
ax.figure.set_size_inches(12,8)
ax.set_title('Nota Média TMDB - Densidade Acumulada', fontsize=15)
ax.set_xlabel('Nota Média', fontsize=13)
ax.set_ylabel('Densidade', fontsize=13)
```

Portanto, um filme com nota média = 7, estaria acima dos 80% dos filmes (estaria no top 20%)

# 02 Explorando as distribuições da amostra
## 2.1 Visualizando outras informações e quantis de forma numérica

```python
tmdb.info()
```

```python
ax = sns.histplot(data=tmdb.query('budget > 0')['budget'], kde=True, stat='density')
ax.figure.set_size_inches(12,8)
ax.set_title('Distribuição de Frequência - Budget (orçamento)', fontsize=15)
ax.set_xlabel('R$ Orçamento', fontsize=13)
ax.set_ylabel('Densidade', fontsize=13)
```

A maior parte dos orçamentos se concentram até 50 milhões 

```python
ax = sns.histplot(data=tmdb.query('vote_count >= 10')['vote_count'], kde=True, stat='density')
ax.figure.set_size_inches(12,8)
ax.set_title('Distribuição de Frequência - # Votos', fontsize=15)
ax.set_xlabel('Quantidade de Votos', fontsize=13)
ax.set_ylabel('Densidade', fontsize=13)
```

A maior parte dos filmes ganham poucos votos, são poucos os filmes com alta quantidade de votos

```python
ax = sns.histplot(data=tmdb['popularity'], kde=True, stat='density')
ax.figure.set_size_inches(12,8)
ax.set_title('Distribuição de Frequência - Popularidade', fontsize=15)
ax.set_xlabel('Popularidade', fontsize=13)
ax.set_ylabel('Densidade', fontsize=13)
```

```python
ax = sns.histplot(data=tmdb.query('runtime > 0')['runtime'].dropna(), kde=True, stat='density')
ax.figure.set_size_inches(12,8)
ax.set_title('Distribuição de Frequência - Runtime (tempo duração)', fontsize=15)
ax.set_xlabel('Tempo duração em minutos', fontsize=13)
ax.set_ylabel('Densidade', fontsize=13)
```

```python
ax = sns.histplot(data=tmdb.query('runtime > 0')['runtime'].dropna(), kde=True, stat='density', cumulative=True)
ax.figure.set_size_inches(12,8)
ax.set_title('Distribuição de Frequência - Runtime (tempo duração)', fontsize=15)
ax.set_xlabel('Tempo duração em minutos', fontsize=13)
ax.set_ylabel('Densidade Acumulada', fontsize=13)
```

Graficamente podemos ver que 80% dos filmes possuem duração inferior a 120min (visualmente)
